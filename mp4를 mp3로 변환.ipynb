```python
# 모듈 설치 : pip install moviepy
from moviepy import VideoFileClip
import os
```

```python
# mp4데이터가 있는 곳 (변환할 데이터)
input_folder = "./mp4/"
# mp3로 저장된 데이터를 받을 곳
output_foler = "./mp3/"
```

```python
for filename in os.listdir(input_folder):
    # 파일이 .mp4로 끝나는지 확인
    if filename.endswith('.mp4'):
        input_path = os.path.join(input_folder, filename)
        # 출력 파일 이름 (확장자만 .mp3로 변경)
        output_filename = filename.replace('.mp4', '.mp3')
        output_path = os.path.join(output_folder, output_filename)

        # 비디오 클립 로드
        video_clip = VideoFileClip(input_path)

        # 오디오 추출 및 저장
        audio_clip = video_clip.audio
        # codec='libmp3lame'는 mp3 인코딩에 필요
        audio_clip.write_audiofile(output_path, codec='libmp3lame')
        

        # 클립 닫기 (메모리 해제)
        video_clip.close()
        audio_clip.close()

print("모든 MP4 파일 변환 완료!")
```
